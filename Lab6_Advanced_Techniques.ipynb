```python
# ADVANCED TECHNIQUES 1 
import numpy as np
try:
    a = np.array([1, 2, 3])
    b = np.array([[1], [2], [3]])
    result = a + b
    print("Broadcasting Result:\n", result)
except ValueError as e:
    print(f"Broadcasting error: {e}")
```

```python
# ADVANCED TECHNIQUES 2
identity = np.eye(3)
matrix = np.array([[2, 3, 4], [1, 0, 6], [7, 8, 9]])
print("Matrix + Identity:\n", matrix + identity)
```

```python
# ADVANCED TECHNIQUES 3
data = np.array([1, 2, np.nan, 4, np.inf])
# This keeps only finite numbers and removes NaNs/Infs
cleaned = data[np.isfinite(data)]
print("Cleaned Data:", cleaned)
```

```python
# ADVANCED TECHNIQUES 4
import pandas as pd
df = pd.DataFrame({
    "Name": ["Jona", "Jones", "Jonalyn", "Jonx"],
    "Score": [85, None, 92, 88]
})
df_cleaned = df.dropna()
print("Cleaned DataFrame:\n", df_cleaned)
```

```python
# ADVANCED TECHNIQUES 5
df = pd.DataFrame({
    "X": [1, 2, 3],
    "Y": [4, 5, 6]
})
df["Log_X"] = np.log(df["X"])
df["Sqrt_Y"] = np.sqrt(df["Y"])
print("Transformed DataFrame:\n", df)
```
